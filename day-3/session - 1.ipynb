```python
wordclean = sorted(list(set([word.strip().lower() for word in open('words.txt' , 'r')])))
```

```python
wordclean
```

```python
def signature(word):
    return ''.join(sorted(word))
```

```python
signature("winemine")
```

```python
def anagrams(mywords):
    return [word for word in wordclean if signature(word) == signature(mywords)]
```

```python
anagrams('cat')
```

```python
%timeit anagrams('cat')
```

```python
words_bysig = {}
for word in wordclean:
    words_bysig[signature(word)].append(word)
```

```python
import collections
words_bysig = collections.defaultdict(list)
for word in wordclean:
    words_bysig[signature(word)].append(word)
words_bysig
```

```python
def anagram_fast(mywors):
    return words_bysig[signature(mywors)]
```

```python
%timeit anagram_fast('cat')
```

```python
wordlen = collections.defaultdict(list)

for word in wordclean:
    wordlen[len(word)].append(word)
wordlen

    
```

```python
worlen = collections.OrderedDict(sorted(wordlen.items()))
worlen
```

```python
anagram_bylength = {}
print(worlen)

for length , words in worlen.items():
       anagram_bylength[length] =  {word: anagram_fast(word) for word in words if len(anagram_fast(words)) > 1}
```

```python
#LOL the commonly used words dont have any anagrams in it.
anagram_bylength
```
